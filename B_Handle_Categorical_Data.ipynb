# Handle Categorical Data

```python
import pandas as pd
 
data = pd.read_csv('..\Linear_Regression\Data\insurance.csv')
display(data.head())
```

```python
=> Here data has 3 categorical features.[sex, smoker, region]
```

## 1. Observe Labels In Each Feature

```python
print(data.sex.unique())
print(data.smoker.unique())
print(data.region.unique())
```

## 2. Label Encoding

```python
1. sex and smoker has two labels only so we can do binary encoding on these features.
2. region has more columns so will do OneHotEncoding for this feature.
```

### 2.1 Encode the sex column

```python
data['sex'] = data['sex'].replace({'female': 0, 'male': 1})
display(data.head())
```

### 2.2 Encode the smoker column

```python
data['smoker'] = data['smoker'].replace({'no': 0, 'yes': 1})
display(data.head())
```

### 2.3 Encode the region column

```python
data = pd.get_dummies(data, columns=['region'], drop_first=True)
display(data)
```

## 3. Observe Data Types 

```python
print(data.dtypes)
```

## 3. Save Data as csv File

```python
1. The data will save in \Linear_Regression\Data\Encoded_Data.csv
2. Saved file name is Encoded_Data.csv
3. This file will use in next process (feature scaling).
```

```python
data.to_csv('..\Linear_Regression\Data\Encoded_Data.csv')
```
