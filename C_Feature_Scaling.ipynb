# Feature Scaling

```python
import pandas as pd

data = pd.read_csv('..\Linear_Regression\Data\Encoded_Data.csv')
display(data.head())
```

```python
data.drop(columns='Unnamed: 0', inplace=True)
display(data.head())
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0.1, 1.1))
data.loc[:, ['age', 'bmi', 'charges']] = scaler.fit_transform(data.loc[:, ['age', 'bmi', 'charges']])
display(data.head())
```

```python
print(data.describe())
```

```python
data.to_csv('..\Linear_Regression\Data\Scaled_Data.csv')
```
